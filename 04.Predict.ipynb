```python
import pickle
def Load_Object(filename):
    with open(filename, 'rb') as file:
        obj = pickle.load(file)
    return obj
```

```python
my_model = Load_Object('food_LogisticRegression.pkl')
count = Load_Object('Count_tfidf.pkl')
```

```python
def xu_ly(text):
  text = convert_emoticons(text, EMOTICONS_EMO)
  text = convert_emoji(text, emojicon_dict)
  text = replace_repeted(text)
  text = convert_teencode(text, teencode_dict)
  text = remove_special_icons(text)
  text = convert_khong_words(text, final_ko_list)
  text = clean_text(text)
  text = word_tokenize(text, format='text')
  return text
```

```python
x_new = 'Đặt đơn bôm đợt 12/12 nên được tặng cái hold cuo khá cute của ShopeeFood. Trà Phúc Long thì mê xưa giờ rồi, trà đậm vị uống chất lượng uống ghiện khó cai. Ly size lớn uống xỉu uo xỉu down vì khá nhiều. Chi nhánh này pha chế khá ổn.'
```

```python
x_new = xu_ly(x_new)
x_new
```

```python
x_new = count.transform(np.array([x_new])).toarray()
x_new
```
